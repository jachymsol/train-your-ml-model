```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
from PIL import Image, ImageEnhance
```

```python
im = Image.open('picture.png')
```

```python
bw = im.convert('L')
```

```python
enhancer = ImageEnhance.Contrast(bw)
contrasted = enhancer.enhance(10)
contrasted
```

```python
array = np.array(contrasted)
array
```

```python
flat_array = array.flatten()
flat_array
```

```python
hist, bins = np.histogram(flat_array, bins=257, range=(-1, 255))
```

```python
import matplotlib.pyplot as plt
```

```python
# Plot the histogram
plt.bar(bins[:-1], hist, width=1)
plt.xlim(min(bins), max(bins))
plt.title('Grayscale Value Distribution')
plt.xlabel('Grayscale Value')
plt.ylabel('Frequency')
plt.show()
```

```python
array = array // 4
```

```python
dark_im = Image.fromarray(array)
dark_im
```

```python
dark_enhancer = ImageEnhance.Contrast(dark_im)
dark_contrasted = dark_enhancer.enhance(50)
dark_contrasted
```
